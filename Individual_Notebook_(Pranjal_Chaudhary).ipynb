## **Adding new features to dataset** 

1) Adding the new column "TotalMinutesSpent" which tells us the total duration for which the user is using the network 

```python
def total_minutes(a,b,c,d):
  return(a+b+c+d)

df["TotalMinutesSpent"]=df.apply(lambda x:total_minutes(x["TotalEveMinutes"],x["TotalDayMinutes"],x["TotalNightMinutes"],x["TotalIntlMinutes"]),axis=1)
df.head(5)
```

2) Adding the new column "ChargePerMinute" which tells how much the user is charged per minute for the different categories.

```python
df["IntlChargePerMinute"]=df["TotalIntlCharge"]/df['TotalIntlMinutes']
df["DayChargePerMinute"]=df["TotalDayCharge"]/df['TotalDayMinutes']
df["NightChargePerMinute"]=df["TotalNightCharge"]/df['TotalNightMinutes']
df["EveChargePerMinute"]=df["TotalEveCharge"]/df['TotalEveMinutes']
df.head()
```

***4) What are the states with high churning rates and what is the distribution among states in terms of high,medium or low churning rates***

```python
#Filtering the state_churn df to make a new dataframe with only values for which churn is true
state_churn_true=state_churn[state_churn["Churn"]==True].copy()

#Finding the mean and standard deviation for the churn count
mean=state_churn_true.describe().transpose()["mean"]
std=state_churn_true.describe().transpose()["std"]

def labeller(a):
  """Labels a given state on the basis of its churn count.
  High_Churn (if churn count is more than the mean)
  Medium_Churn (if churn count is less than mean but more than the mean minus the standard deviation)
  Low Churn (if churn count does not satisfy the above conditions)"""
  if a>mean['Churn_count']:
    return "High_Churn"
  elif a>mean["Churn_count"]-std["Churn_count"]:
    return "Medium_Churn"
  else:
    return "Low_Churn"

#Using apply function to provide labels to state on the basis of above mentioned criteria
state_churn_true["label"]=state_churn_true.loc[:,"Churn_count"].apply(labeller)

#Making the pie plot from the data
labels=list(state_churn_true.groupby(by=["label"])["label"].count().index)
values=list(state_churn_true.groupby(by=["label"])["label"].count().values)
colors = sns.color_palette('bright')[7:10]
explode=(0.1,0.1,0.1)
plt.pie(values, labels = labels, colors = colors, autopct='%0.1f%%',shadow=True,explode=explode,textprops={'fontsize': 13})
plt.title("Distribution of Churning Rates", size=15,pad=25)
#plt.savefig('High_Churn_States.png')
#files.download('High_Churn_States.png')
plt.show()
```

*  Nearly 42% of the states have number of churned users higher than the average churn for the states.

***6) Determining number of churned and not churned users on the basis of Customer Service Calls***

```python
#Creating a dataframe to calculate the churned users for different number of Customer Service Calls
churn_true=pd.DataFrame(df[df["Churn"]==True].groupby("CustomerServiceCalls",as_index=False)["Churn"].count())
churn_true.rename(columns={"Churn":"ChurnCount"},inplace=True)

#Creating a dataframe to calculate the NonChurned users for different number of Customer Service Calls
churn_false=pd.DataFrame(df[df["Churn"]==False].groupby("CustomerServiceCalls",as_index=False)["Churn"].count())
churn_false.rename(columns={"Churn":"NotChurnCount"},inplace=True)

#Calculating the total number of churned and non churned users
total_churned_users=churn_true["ChurnCount"].sum()
total_not_churned_users=churn_false["NotChurnCount"].sum()

#Calculating the percentage of churned and non-churned users 
churn_true["Churned"]=(churn_true["ChurnCount"]/(total_churned_users))*100
churn_false["NotChurned"]=(churn_false["NotChurnCount"]/(total_not_churned_users))*100
```

```python
#Creating a new dataframe by merging churn_true and churn_false dataframes
new_df=pd.merge(churn_false,churn_true,on="CustomerServiceCalls")
df_churn=new_df[["CustomerServiceCalls","Churned","NotChurned"]]
#Setting the figure size parameters
plt.rcParams["figure.figsize"] = [10, 7]

#Using the subplots method to create subplots
f,axes=plt.subplots(1,2,figsize=(15,6))

#Creating the first visualisation
v1=sns.lineplot(data=df_churn, x="CustomerServiceCalls",y="Churned",ax=axes[0])
axes[0].set_ylabel("Number of Churned Users",size=15,labelpad=10)
axes[0].set_xlabel("Number of customer service calls",size=15,labelpad=20)

#Creating the second visualisation
v2=sns.lineplot(data=df_churn, x="CustomerServiceCalls",y="NotChurned",ax=axes[1],markers=True,dashes=False,palette="flare")
axes[1].set_ylabel("Number of NonChurned Users",size=15,labelpad=10)
axes[1].set_xlabel("Number of customer service calls",size=15,labelpad=20)

plt.xticks(rotation=0)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.suptitle("Number of churn and non_churned users vs number of customer service calls",size=17)
#plt.savefig('customerservicecallchurn.png')  # to save and download the files
#files.download('customerservicecallchurn.png')
plt.show()
```

*  The user retention looks to decrease with increasing number of customer service calls while the churning rate seems to increase.

***8) Is there any relation between the percentage of churned users vs their total day charge ?***

```python
#Creating a dataframe having info abour churned users who spend more than average on their day time calls
churn_more_than_average=df[(df.Churn==True) & (df.TotalDayCharge >= df.TotalDayCharge.mean()) ]["Churn"].count()

#Creating a dataframe having info abour non-churned users who spend more than average on their day time calls
nonchurn_more_than_average=df[(df.Churn==False) & (df.TotalDayCharge >= df.TotalDayCharge.mean()) ]["Churn"].count()

#Calculating the total number of users satisfying the criteria
more_than_average=(churn_more_than_average+nonchurn_more_than_average)

#Converting the values to percentages
churn_more_than_average=(churn_more_than_average*100)/more_than_average
nonchurn_more_than_average=(nonchurn_more_than_average*100)/more_than_average

#Performing ths similar actions as done above for users who spent less than average on their day time calls
churn_less_than_average=df[(df.Churn==True) & (df.TotalDayCharge < df.TotalDayCharge.mean()) ]["Churn"].count()
nonchurn_less_than_average=df[(df.Churn==False) & (df.TotalDayCharge < df.TotalDayCharge.mean()) ]["Churn"].count()
less_than_average=(churn_less_than_average+nonchurn_less_than_average)
churn_less_than_average=(churn_less_than_average*100)/less_than_average
nonchurn_less_than_average=(nonchurn_less_than_average*100)/less_than_average

#Making a new dataframe to store all the above calculated values
df_day_churn=pd.DataFrame({"CustomerCategory":["day_charges_more_than_avg","day_charges_more_than_avg","day_charges_less_than_avg","day_charges_less_than_avg"],
                  "Churn":[True,False,True,False],
                  "PercentOfUsers":[churn_more_than_average,nonchurn_more_than_average,churn_less_than_average,nonchurn_less_than_average]})

#Plotting the information
plt.rcParams["figure.figsize"] = [11, 8]
vis8=sns.barplot(data=df_day_churn, x="CustomerCategory",y="PercentOfUsers",dodge=False,hue="Churn")
plt.legend(prop={"size":13},loc='upper left')
#plt.savefig('day_charge_churn.png')
#files.download('day_charge_churn.png')
plt.show()
df_day_churn
```

*  The percentage of churned users for user spending more than the average day charges is nearly 6% more than that of users who spend less than average day charges. 

***9) What is the relation of churn and non churn percentages for users who have or have not opted for voice mail ?***

```python
#Creating a dataframe with users who have and have not opted for voice mail plans and thier churn value
voice_mail_churn=df.groupby(by=["VoiceMailPlan","Churn"],as_index=False).agg(Churn_count=('Churn','count'))
not_have_voicemail=voice_mail_churn.groupby("VoiceMailPlan")["Churn_count"].sum()["No"]
have_voicemail=voice_mail_churn.groupby("VoiceMailPlan")["Churn_count"].sum()["Yes"]
voice_mail_churn

#defining label function to provide a new column which converts the values to their percentages respectively
def label(a,b):
  if a=="No":
    return((b*100)/not_have_voicemail)
  else:
    return((b*100)/have_voicemail)

voice_mail_churn["Churn_Percent"]=voice_mail_churn.apply(lambda x:label(x["VoiceMailPlan"],x["Churn_count"]),axis=1)
voice_mail_churn
```

```python
#Setting up the parameters for the figure
plt.rcParams["figure.figsize"] = [7, 7]
sns.set_style('darkgrid')

#Creating subplot objects 
f,axes=plt.subplots(1,2,figsize=(15,6))

#Creating the first visualisation
vis1=sns.barplot(x="VoiceMailPlan",y="Churn_count",data=voice_mail_churn,hue="Churn",dodge=False,ax=axes[0])
axes[0].set_ylabel("Number of Users",size=15,labelpad=10)
axes[0].set_xlabel("Does User have Voice Plan",size=18,labelpad=20)
axes[0].legend(prop={'size':10})

#Creating the second visualisation
vis2=sns.barplot(x="VoiceMailPlan",y="Churn_Percent",data=voice_mail_churn,hue="Churn",dodge=False,ax=axes[1])
plt.xticks(rotation=90)
plt.title("Distribution of churn and non_churned users who have or have not opted for voice mail plan",size=17,loc='right')
axes[1].set_ylabel("Percentage of Users",size=15,labelpad=10)
axes[1].set_xlabel("Does User have Voice Plan",size=18,labelpad=15)
axes[1].legend(prop={"size":10},loc='upper left')

#plt.savefig('churn_voice_mail.png')  
#files.download('churn_voice_mail.png')

plt.show()
```

* Nearly twice the percentage of users have churned who did not opt for the voice plan when compared to the percent of users who opted for the plans.
